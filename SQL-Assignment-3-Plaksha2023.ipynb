```python
# sqlite3 package comes with the Python installation
import sqlite3
import pandas as pd
```

```python
#We create a table called students. It has the columns: 'student_id', 'name', 'major', 'gpa' and 'enrollment_date' 
connection = sqlite3.connect('company.db')
cursor = connection.cursor()
sql_command = '''
CREATE TABLE students AS
    SELECT 1 AS student_id, "John" AS name, "Computer Science" AS major, 3.5 AS gpa, "01-01-2022" AS enrollment_date UNION
    SELECT 2, "Jane", "Physics", 3.8, "01-02-2022" UNION
    SELECT 3, "Bob", "Engineering", 3.0, "01-03-2022" UNION
    SELECT 4, "Samantha", "Physics", 3.9, "01-04-2022" UNION
    SELECT 5, "James", "Engineering", 3.7, "01-05-2022" UNION
    SELECT 6, "Emily", "Computer Science", 3.6, "01-06-2022" UNION
    SELECT 7, "Michael", "Computer Science", 3.2, "01-07-2022" UNION
    SELECT 8, "Jessica", "Engineering", 3.8, "01-08-2022" UNION
    SELECT 9, "Jacob", "Physics", 3.4, "01-09-2022" UNION
    SELECT 10, "Ashley", "Physics", 3.9, "01-10-2022"
'''
# cursor.execute(sql_command).fetchall()
```

```python
#We SELECT all records in the table.
sql_command= '''
SELECT * 
FROM students
'''
pd.read_sql_query(sql_command,connection)
```

```python
#We SELECT students whose major is "Computer Science".
sql_command='''
SELECT * 
FROM students
WHERE major='Computer Science'
'''
pd.read_sql_query(sql_command,connection)
```

```python
#We SELECT all unique majors (use SELECT DISTINCT) and order them by name, descending order (i.e. Physics first).
sql_command='''
SELECT DISTINCT major
FROM students
ORDER BY major DESC;
'''
pd.read_sql_query(sql_command,connection)
```

```python
#We SELECT all students that have an 'e' in their name and order them by gpa in ascending order.
sql_command='''
SELECT * 
FROM students 
WHERE name LIKE '%e%'
ORDER BY gpa;
'''
pd.read_sql_query(sql_command,connection)
```

```python
#We create a new table called courses, which indicates the courses taken by the students.
sql_command='''
CREATE TABLE courses AS
    SELECT 1 AS course_id, "Python programming" AS course_name, 1 AS student_id, "A" AS grade UNION
    SELECT 2, "Data Structures", 2, "B" UNION
    SELECT 3, "Database Systems", 3, "B" UNION
    SELECT 1, "Python programming", 4, "A" UNION
    SELECT 4, "Quantum Mechanics", 5, "C" UNION
    SELECT 1, "Python programming", 6, "F" UNION
    SELECT 2, "Data Structures", 7, "C" UNION
    SELECT 3, "Database Systems", 8, "A" UNION
    SELECT 4, "Quantum Mechanics", 9, "A" UNION
    SELECT 2, "Data Structures", 10, "F";
'''
# cursor.execute(sql_command).fetchall() 
```

```python
#We COUNT the number of unique courses.
sql_command='''
SELECT COUNT (DISTINCT course_name) AS no_of_distinct_courses
FROM courses;
'''
print("The number of unique courses is",cursor.execute(sql_command).fetchall())
```

```python
#We JOIN the tables students and courses and COUNT the number of students with the major Computer Science taking the course Python programming.
sql_command='''
SELECT COUNT(*) AS comp_sci_major_with_python
FROM students
INNER JOIN courses
USING (student_id)
WHERE major = 'Computer Science' AND course_name='Python programming';
'''
pd.read_sql_query(sql_command,connection)
```

```python
#We JOIN the tables students and courses and select the students who have grades higher than "C", only show their name, major, gpa, course_name and grade.
sql_command='''
SELECT *
FROM students
INNER JOIN courses
USING (student_id)
WHERE grade<'C';
'''
pd.read_sql_query(sql_command,connection)
```

```python
#We Find the average gpa of all students.
sql_command='''
SELECT AVG(gpa) AS avg_gpa
FROM students 
'''
pd.read_sql_query(sql_command,connection)
```

```python
#We SELECT the student with the maximum gpa, display only their student_id, major and gpa
sql_command='''
SELECT student_id,major,gpa
FROM students
WHERE gpa IN (SELECT MAX(gpa)
             FROM students)
'''
pd.read_sql_query(sql_command,connection)
```

```python
#We SELECT the student with the minimum gpa, display only their student_id, major and gpa
sql_command='''
SELECT student_id,major,gpa
FROM students
WHERE gpa IN (SELECT MIN(gpa)
             FROM students)
'''
pd.read_sql_query(sql_command,connection)
```

```python
#We SELECT the students with a gpa greater than 3.6 in the majors of "Physics" and "Engineering", display only their student_id, major and gpa
sql_command='''
SELECT student_id,major,gpa
FROM students
WHERE gpa>3.6 AND major IN ('Physics','Engineering');
'''
pd.read_sql_query(sql_command,connection)
```

```python
#We Group the students by their major and retrieve the average grade of each major
sql_command='''
SELECT major,AVG(gpa)
FROM students
GROUP BY major;
'''
pd.read_sql_query(sql_command,connection)
```

```python
#We SELECT the top 2 students with the highest GPA in each major and order the results by major in ascending order, then by GPA in descending order
sql_command='''
SELECT student_id,name,major,gpa,enrollment_date
FROM (SELECT student_id,name,major,gpa,enrollment_date,
     ROW_NUMBER() OVER (PARTITION BY major ORDER BY gpa DESC) AS rank_within_major
     FROM students
     )
WHERE rank_within_major< 3
ORDER BY major,gpa DESC;
'''
pd.read_sql_query(sql_command,connection)
```
